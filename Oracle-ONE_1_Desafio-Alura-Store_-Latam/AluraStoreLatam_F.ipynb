### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# 01.INGRESO TOTAL POR TIENDA

## 1.1 Calcular el Ingreso Total de Cada Tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("="*60)
print("Desafio 1: INGRESO TOTAL POR TIENDA")
print("="*60)
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f}")
print("="*60)
```

```python
## 1.2 Visualización de Ingreso Total de Cada Tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['blue', 'green', 'orange', 'red'])
plt.title('Ingreso Total por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tiendas')
plt.ylabel('Ingreso Total ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

# Agregar etiquetas de valores
for i, v in enumerate(ingresos):
    plt.text(i, v, f'${v:,.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# 02. VENTAS POR CATEGORÍA

## 2.1 Calcular la cantidad de productos vendidos por categoría del producto en cada tienda

ventas_categoria_t1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_t4 = tienda4['Categoría del Producto'].value_counts()

print("="*60)
print("RETO 2: VENTAS POR CATEGORÍA")
print("="*60)
print("\n=== TIENDA 1 - Ventas por Categoría ===")
print(ventas_categoria_t1)
print("\n=== TIENDA 2 - Ventas por Categoría ===")
print(ventas_categoria_t2)
print("\n=== TIENDA 3 - Ventas por Categoría ===")
print(ventas_categoria_t3)
print("\n=== TIENDA 4 - Ventas por Categoría ===")
print(ventas_categoria_t4)
print("="*60)
```

```python
## 2.2 Visualización de ventas por categoría para cada tienda
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

ventas_categoria_t1.plot(kind='barh', ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Tienda 1 - Ventas por Categoría', fontweight='bold')
axes[0, 0].set_xlabel('Cantidad de Ventas')
axes[0, 0].set_ylabel('Categoría')

ventas_categoria_t2.plot(kind='barh', ax=axes[0, 1], color='green')
axes[0, 1].set_title('Tienda 2 - Ventas por Categoría', fontweight='bold')
axes[0, 1].set_xlabel('Cantidad de Ventas')
axes[0, 1].set_ylabel('Categoría')

ventas_categoria_t3.plot(kind='barh', ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Tienda 3 - Ventas por Categoría', fontweight='bold')
axes[1, 0].set_xlabel('Cantidad de Ventas')
axes[1, 0].set_ylabel('Categoría')

ventas_categoria_t4.plot(kind='barh', ax=axes[1, 1], color='red')
axes[1, 1].set_title('Tienda 4 - Ventas por Categoría', fontweight='bold')
axes[1, 1].set_xlabel('Cantidad de Ventas')
axes[1, 1].set_ylabel('Categoría')

plt.tight_layout()
plt.show()
```

```python
## 6.3 Gráfico de Caja (BoxPlot): Distribución de Precios por Tienda

### 6.3.1 Preparar datos para boxplot
data_precios = [tienda['Precio'], tienda2['Precio'], tienda3['Precio'], tienda4['Precio']]
labels_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(12, 7))
bp = plt.boxplot(data_precios, labels=labels_tiendas, patch_artist=True,
                 boxprops=dict(facecolor='lightblue', color='blue'),
                 medianprops=dict(color='red', linewidth=2),
                 whiskerprops=dict(color='blue'),
                 capprops=dict(color='blue'))

### 6.3.2 Colorear cada caja
colors = ['blue', 'green', 'orange', 'red']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.5)

plt.title('Distribución de Precios por Tienda (BoxPlot)', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Precio ($)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("="*80)
print("INSIGHT: El BoxPlot muestra la mediana, cuartiles y valores atípicos")
print("de los precios, permitiendo comparar la dispersión de precios entre tiendas.")
print("="*80)
```

```python
## 6.4 Histograma: Distribución de Calificaciones (Todas las Tiendas)

### 6.4.1 Combinar calificaciones de todas las tiendas
import pandas as pd
todas_calificaciones = pd.concat([
    tienda['Calificación'],
    tienda2['Calificación'],
    tienda3['Calificación'],
    tienda4['Calificación']
])

plt.figure(figsize=(12, 6))
plt.hist(todas_calificaciones, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(todas_calificaciones.mean(), color='red', linestyle='--',
            linewidth=2, label=f'Media: {todas_calificaciones.mean():.2f}')
plt.axvline(todas_calificaciones.median(), color='green', linestyle='--',
            linewidth=2, label=f'Mediana: {todas_calificaciones.median():.2f}')

plt.title('Distribución de Calificaciones de Clientes (Todas las Tiendas)',
          fontsize=16, fontweight='bold')
plt.xlabel('Calificación', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("="*80)
print("INSIGHT: El histograma muestra la distribución de las calificaciones,")
print("revelando la satisfacción general de los clientes.")
print("="*80)
```

```python
## 6.5 Gráfico de Líneas: Comparativa de Métricas Normalizadas por Tienda

### 6.5.1 Normalizar métricas para comparación en la misma escala (0-100)
def normalizar_0_100(valores):
    """Normaliza valores a escala 0-100"""
    min_val = min(valores)
    max_val = max(valores)
    if max_val == min_val:
        return [50] * len(valores)
    return [(v - min_val) / (max_val - min_val) * 100 for v in valores]

### 6.5.2 Preparar datos normalizados
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos_norm = normalizar_0_100(ingresos)
calificaciones_norm = normalizar_0_100(calificaciones)
envios_norm = normalizar_0_100(envios)
### 6.5.3  Invertir envíos (menor es mejor)
envios_norm_inv = [100 - v for v in envios_norm]

### 6.5.4 Crear gráfico de líneas
plt.figure(figsize=(12, 7))
plt.plot(tiendas_nombres, ingresos_norm, marker='o', linewidth=2,
         markersize=10, label='Ingresos (normalizado)', color='blue')
plt.plot(tiendas_nombres, calificaciones_norm, marker='s', linewidth=2,
         markersize=10, label='Calificación (normalizado)', color='green')
plt.plot(tiendas_nombres, envios_norm_inv, marker='^', linewidth=2,
         markersize=10, label='Calidad Envío (normalizado)', color='orange')

plt.title('Comparativa de Métricas Clave por Tienda (Escala 0-100)',
          fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Puntuación Normalizada (0-100)', fontsize=12)
plt.ylim(0, 105)
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("="*80)
print("INSIGHT: Este gráfico de líneas permite comparar el desempeño relativo")
print("de cada tienda en múltiples métricas simultáneamente.")
print("="*80)
```

Para determinar qué tienda debe vender el Sr. Juan, se establecieron los siguientes criterios ponderados:

### 3.1 Criterios de Evaluación

| Criterio | Peso | Justificación |
|----------|------|---------------|
| **Ingresos Totales** | 30% | Refleja la capacidad de generación de valor |
| **Calificación del Cliente** | 25% | Indica satisfacción y lealtad de clientes |
| **Eficiencia Operativa (Envío)** | 20% | Muestra optimización de costos |
| **Diversificación** | 15% | Reduce riesgo de dependencia de productos |

### 3.2 Perspectiva del Vendedor (Sr. Juan)

Desde el punto de vista del Sr. Juan, **vender la tienda menos atractiva** podría implicar:

- Deshacerse de activos de bajo rendimiento
- Liberar capital para reinversión en tiendas más rentables
- Reducir complejidad operativa

**Alternativamente**, **vender la tienda más exitosa** podría significar:

- Maximizar el precio de venta (mayor valoración)
- Obtener liquidez inmediata significativa
- Reducir exposición en mercados saturados
- Aprovechar momento de alto valor de mercado

```python
## CÁLCULO DE SCORE PONDERADO POR TIENDA

import pandas as pd

# Definir pesos de criterios
pesos = {
    'Ingresos': 0.30,
    'Calificación': 0.25,
    'Eficiencia_Envío': 0.20,
    'Diversificación': 0.15,
    'Cobertura_Geo': 0.10
}

# Normalizar métricas a escala 0-100
def normalize_100(values, invertir=False):
    min_val = min(values)
    max_val = max(values)
    if max_val == min_val:
        return [50] * len(values)
    normalized = [(v - min_val) / (max_val - min_val) * 100 for v in values]
    if invertir:
        normalized = [100 - v for v in normalized]
    return normalized

# Preparar datos para scoring
tiendas_scoring = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Score_Ingresos': normalize_100([ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]),
    'Score_Calificacion': normalize_100([calificacion_promedio_t1, calificacion_promedio_t2,
                                         calificacion_promedio_t3, calificacion_promedio_t4]),
    'Score_Eficiencia': normalize_100([envio_promedio_t1, envio_promedio_t2,
                                       envio_promedio_t3, envio_promedio_t4], invertir=True),
    'Score_Diversificacion': normalize_100([tienda['Producto'].nunique(), tienda2['Producto'].nunique(),
                                           tienda3['Producto'].nunique(), tienda4['Producto'].nunique()]),
    'Score_Cobertura': normalize_100([tienda['Lugar de Compra'].nunique(),
                                      tienda2['Lugar de Compra'].nunique(),
                                      tienda3['Lugar de Compra'].nunique(),
                                      tienda4['Lugar de Compra'].nunique()])
})

# Calcular score total ponderado
tiendas_scoring['Score_Total'] = (
    tiendas_scoring['Score_Ingresos'] * pesos['Ingresos'] +
    tiendas_scoring['Score_Calificacion'] * pesos['Calificación'] +
    tiendas_scoring['Score_Eficiencia'] * pesos['Eficiencia_Envío'] +
    tiendas_scoring['Score_Diversificacion'] * pesos['Diversificación'] +
    tiendas_scoring['Score_Cobertura'] * pesos['Cobertura_Geo']
).round(2)

# Ordenar por score total
tiendas_scoring = tiendas_scoring.sort_values('Score_Total', ascending=False)

print("\n" + "="*100)
print(" "*30 + "MATRIZ DE PUNTUACIÓN PONDERADA")
print("="*100)
print("\nPesos aplicados:")
print(f"  • Ingresos: {pesos['Ingresos']*100:.0f}%")
print(f"  • Calificación: {pesos['Calificación']*100:.0f}%")
print(f"  • Eficiencia Envío: {pesos['Eficiencia_Envío']*100:.0f}%")
print(f"  • Diversificación: {pesos['Diversificación']*100:.0f}%")
print(f"  • Cobertura Geográfica: {pesos['Cobertura_Geo']*100:.0f}%")
print("\n" + "-"*100)
print(tiendas_scoring[['Tienda', 'Score_Ingresos', 'Score_Calificacion', 'Score_Eficiencia',
                        'Score_Diversificacion', 'Score_Cobertura', 'Score_Total']].to_string(index=False))
print("="*100)

# Identificar mejor y peor tienda
mejor_tienda = tiendas_scoring.iloc[0]['Tienda']
peor_tienda = tiendas_scoring.iloc[-1]['Tienda']
mejor_score = tiendas_scoring.iloc[0]['Score_Total']
peor_score = tiendas_scoring.iloc[-1]['Score_Total']

print(f"\n🥇 TIENDA DE MAYOR VALOR: {mejor_tienda} (Score: {mejor_score}/100)")
print(f"🥉 TIENDA DE MENOR VALOR: {peor_tienda} (Score: {peor_score}/100)")
print("="*100)
```
